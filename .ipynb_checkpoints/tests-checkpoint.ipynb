```python
import os
from zipfile import ZipFile
import pandas as pd
```

```python
# build the index.html

files = [f for f in os.listdir('.') if 
         (os.path.isfile(f) and f[-4:]=='html')]
with open('index.html', 'w') as f:
    text = """<html>
    <h1>Welcome to my wepage</h1>
    
    <body><p>File directory</p><ul> """
    
    for file in files:
        text += "<li><a href=" + file + ">" + file + "</a></li>"
    text += "</ul> \n </body> \n </html>"
    f.write(text)
```

```python
counts = {}
def mydec(fn):
    def wrapper():
        if fn.__name__ not in counts:
            counts[fn.__name__] = 0
        counts[fn.__name__] += 1
        return fn()
    return wrapper

@mydec
def g():
    print('g')
    
@mydec
def f():
    print('f')
    
f()
g()
g()
```
